```python
with NWBHDF5IO("sub-489_ses-20220708.nwb", "r") as io:
    nwb0 = io.read()
    trial_table = nwb0.acquisition['trial'].data[:]
    #print(trial_table)
    #print("All trial types:", nwb0.trials['trial_type'].data[:])
    #print(nwb0)
    #start_timeprint(nwb0.acquisition.keys())
    #print(len(nwb0.units))
    #locations=nwb0.electrodes['location'][:]
    #print(set(locations))
    #print(nwb0.electrode_groups.keys()) empty
    
    #units= nwb0.units
    #first_spikes = units.get_unit_spike_times(0)
    #print(first_spikes[-20:])
    lateral_distance = nwb0.acquisition['distance_lat'].data[:]
    time = nwb0.acquisition['ts_local'].data[:]
    trial_numbers = nwb0.acquisition['trial'].data[:]

    unique_trial_nums = np.sort(np.unique(trial_numbers))

    trial_segments = {}
    for trial_num in unique_trial_nums:
        mask = trial_numbers == trial_num
        trial_segments[int(trial_num)] = {
            'time': time[mask],
            'lat_dist': lateral_distance[mask]
        }
    output_dir = "trial_plots"
    os.makedirs(output_dir, exist_ok=True)  # Create folder if it doesn't exist
    for trial_num in unique_trial_nums:  # Iterate in order (1, 2, 3, ...)
        seg = trial_segments[int(trial_num)]

        plt.figure(figsize=(10, 4))
        plt.plot(seg['time'], seg['lat_dist'], color='blue')
        plt.axhline(0, color='black', linestyle='--', label='Middle Line (0 mm)')
        plt.xlabel('Time (s)')
        plt.ylabel('Lateral Distance (mm)')
        plt.title(f'Trial {trial_num}')
        plt.legend()
        plt.grid(True)

    # Save and close (critical to avoid memory issues)
        plt.savefig(f"{output_dir}/trial_{trial_num}.png", bbox_inches='tight')
        plt.close()  # Prevents memory overload with 1600+ plots
        
    #segmentation regarding trial type
    trial_type_map = {}
    for trial_num in unique_trial_nums:
    # Assume trial_num corresponds to a filename like 'OL_L_1.dat'
    # This is a placeholder—you need to replace with your actual trial naming logic.
        if trial_num % 2 == 0:  # Example: even numbers = OL_L
            trial_type_map[trial_num] = 'OL_L_1.dat'
        else:  # odd numbers = Straight
            trial_type_map[trial_num] = 'Straight 100cm A.dat'
    trial_data = {
    'Straight': [],
    'OL_L': []
    }

    for trial_num, seg in trial_segments.items():
        trial_type = trial_type_map.get(trial_num, "Unknown")
        if 'Straight' in trial_type:
            trial_data['Straight'].append(seg)
        elif 'OL_L' in trial_type:
            trial_data['OL_L'].append(seg)

    


        #print(f"Saved {len(unique_trial_nums)} plots to '{output_dir}/'")
    
    
with NWBHDF5IO("sub-503_ses-20220902.nwb", "r") as io1:
    nwb1 = io1.read()
    #print(nwb1)
    #units= nwb1.units
    #first_spikes = units.get_unit_spike_times(0)
    #print(first_spikes[-20:])
#print("===Top-level fields===")
#print(list(read_nwbfile.fields.keys()))
#print(list(read_nwbfile1.fields.keys()))
    print(nwb1)
    print("All trial types:", nwb1.trials['trial_type'].data[:])
    trial_table = nwb1.acquisition['trial'].data[:]
    print(nwb1.trials['start time'][:])
    #print(nwb1.trials.to_dataframe())
```
